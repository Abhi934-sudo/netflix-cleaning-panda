```python
import pandas as pd
```

```python

df= pd.read_csv("netflixmovies.csv")
df

```

```python
df.info()
```

```python
df.columns = df.columns.str.strip()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'],errors="coerce")
```

```python
duration_split = df['duration'].str.extract(r'(\d+)\s*(\w+)')
df['duration_num'] = pd.to_numeric(duration_split[0], errors='coerce')
df['duration_unit'] = duration_split[1]
```

```python
df['duration_unit'] = df['duration_unit'].replace({'seasons': 'season'})
```

```python
df['duration_num'] = df.groupby('duration_unit')['duration_num'].transform(lambda x: x.fillna(x.median()))
```

```python
df.head()
```

```python
df = df.drop(columns=['duration'])
```

```python
df.head()
```

```python
df.tail()
```

```python
df['rating_missing'] = df['rating'].isna().astype(int)
df['rating'] = df['rating'].fillna('Unknown')
```

```python
df.duplicated().sum()
```

```python
median_date = df['date_added'].median()
df['date_added'] = df['date_added'].fillna(median_date)
```

```python
df.isnull().sum()
```

```python
df['country_missing'] = df['country'].isna().astype(int)
df['country'] = df['country'].fillna("Unknown")
```

```python
df.isnull().sum()
```

```python
df['director_missing'] = df['director'].isna().astype(int)
df['cast_missing'] = df['cast'].isna().astype(int)
```

```python
df['director'] = df['director'].fillna("Unknown")
df['cast'] = df['cast'].fillna("Unknown")
```
